### Problem 1)

```python
a = int(input("Enter an integer:"))
def chPerf(a):
    sum = 0
    for i in range(1,a):
        if(a%i == 0 ):
            sum += i
    if(a == sum):
        print("Your number is müko")
    else:
        print("Bizimle değılsiniz.")
    
print(chPerf(a))
```

### Problem 2)

```python
def FindEbob1(sayi1,sayi2): 
    ebob=1
    i = 1
    while(i <= sayi1 and i <= sayi2):
        if(not(sayi1%i) and not (sayi2%i)):
            ebob = i
        i+=1
    return ebob
sayi1 = 24
sayi2 = 32
print(FindEbob1(sayi1,sayi2))
```

### Problem 3)

```python
birler =  ["","Bir","İki","Üç","Dört","Beş","Altı","Yedi","Sekiz","Dokuz"]
onlar = ["","On","Yirmi","Otuz","Kırk","Elli","Altmış","Yetmiş","Seksen","Doksan"]

def Talk(x):
    xx= x % 10 
    yy= x // 10
    return onlar[yy] + " " + birler[xx]

print(Talk(10))
```

### Problem 5)

```python
def pisagor_bulma():
    
    pisagor_listesi = list()
    for i in range(1,101):
        for j in range(1,101):

            c = (i ** 2 + j ** 2) ** 0.5

            if (c == int(c) ):
                pisagor_listesi.append((i,j,int(c)))

    return pisagor_listesi

for i in pisagor_bulma():
    print(i)
    
```
